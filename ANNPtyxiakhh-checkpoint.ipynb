```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.datasets import load_iris
import graphviz
```

```python
data = load_iris()
X, y = data.data, data.target
```

```python
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X, y)
```

```python
dot_data = export_graphviz(
    clf,
    out_file=None,  # Εξαγωγή ως string
    feature_names=data.feature_names,  # Ονόματα χαρακτηριστικών
    class_names=data.target_names,  # Ονόματα κατηγοριών
    filled=True,  # Χρώματα για κόμβους
    rounded=True,  # Στρογγυλεμένα πλαίσια
    special_characters=True  # Χρήση ειδικών χαρακτήρων
)

# Δημιουργία του γραφήματος
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # Αποθήκευση ως αρχείο PDF (προαιρετικό)
graph  # Εμφάνιση στο Jupyter Notebook
```
